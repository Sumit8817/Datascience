```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df1=pd.read_csv('Universities.csv')
```

```python
df1
```

```python
plt.hist(df1['SAT'])
plt.xlabel('satscore')
plt.ylabel('frequency')
plt.show()
```

```python
sns.histplot(df1['SAT'])
plt.xlabel('satscore')
plt.show()
```

```python
plt.plot(df1['SAT'], df1['Accept'],'bo')
plt.show()
```

```python
sns.scatterplot(x='SAT', y='Top10', data=df1, marker='s', color='red')
plt.xlabel('SAT Score')
plt.ylabel('Top 10% ')
plt.title('SAT Score vs. Top 10%')
plt.show()
```

```python
sns.boxplot(x='Top10', y='Accept', data=df1)
plt.show()
```
